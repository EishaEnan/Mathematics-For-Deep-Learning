```python
import numpy as np
import matplotlib.pyplot as plt
```

### Probabilities of getting x number heads from 3 coinf flips 

```python

# Sample size
N = 1000000

# Number of flips
M = 3

# Possible heads 0, 1, 2, 3 -> 4
heads = np.zeros(M+1)

for i in range(N):
    flips = np.random.randint(0,2, M)
    h, _ = np.bincount(flips, minlength=2) # head counts
    heads[h] += 1 

prob = heads / N

print(f"Probabilities: {np.array2string(prob)}")
```

### Birthday Paradox
How many people do we need for more than 50% chance of two people shared the same birthday?

```python
# Number of comparison needed so that the probability of not sharing birthday is < 0.5
for n in range(365):
    if ((364/365)**n < 0.5):
        print(n)
        break
```

```python
# Probability of two random people sharing birthday
match = 0
N = 100000
for i in range(N):
    a = np.random.randint(0, 364)
    b = np.random.randint(0, 364)
    if (a == b):
        match += 1

print(f"Probability of a random match = {match/N}")
```

```python
# Run simulation, N times, number of people ranging from 2 to 30 and plot the probability of two people sharing birthday
N = 10000
m_values = []
probabilities = []

# Collecting data
for m in range(2, 31):
    matches = 0
    for n in range(N):
        match = 0
        b = np.random.randint(0, 364, m)
        for i in range(m):
            for j in range(m):
                if (i != j) and (b[i] == b[j]):
                    match += 1
        if match != 0:
            matches += 1
    probability = matches / N
    m_values.append(m)
    probabilities.append(probability)

# Plotting m vs p using matplotlib
plt.plot(m_values, probabilities, marker='o', linestyle='-', color='b')
plt.xlabel('Number of people (m)')
plt.ylabel('Probability of at least one shared birthday (p)')
plt.title('Birthday Paradox: Number of People vs Probability of Shared Birthday')
plt.grid(True)
plt.show()
```

#### 4 out of 50 people have been to Boston during the fall. Simulate the probability of  randomly selecting 3 people and none of them have been to Boston in the fall.

```python
nb = 0
N = 100000
for i in range(N):
    s = np.random.randint(0, 50, 3)
    fail = False
    for t in range(3):
        if (s[t] < 4): # Assuming person 0, 1, 2, and 3 have been to Boston in the fall
            fail = True
    if (not fail):
        nb += 1

print(f"No Boston in the fall = {nb/N}")
```
